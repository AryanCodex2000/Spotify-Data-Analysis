```python
sample_data = sp_tracks[['energy', 'loudness']].sample(frac=0.002)
```

```python
sample_data.shape, sp_tracks.shape
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x = sample_data['energy'], y = sample_data['loudness'], edgecolors='k')
plt.title('Loudness vs Energy')
plt.xlabel('Loudness')
plt.ylabel('Energy')
plt.show()
```

```python
sns.regplot(data=sample_data.sample(frac=0.05), y='loudness', x='energy', scatter_kws={"color": "blue"}, line_kws={"color": "red"}).set(title='Regression Plot - Loudness vs Energy Correlation')
```
